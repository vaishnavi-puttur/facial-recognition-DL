```
import cv2
```

```
import dlib
```

```
img = cv2.imread("ironman.jpg")
```

```
plt.imshow(img)
```

```
type(img)
```

```
img.shape
```

```
out = requests.get("https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml")

with open("haarcascade_frontalface_default.xml", "wb") as f:
  f.write(out.content)
```

```
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```
faces = classifier.detectMultiScale(img)

for face in faces:
  x, y, w, h = face 
  img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 4)
```

```
plt.imshow(img)
```

```
import dlib
```

```
out = requests.get("https://raw.githubusercontent.com/italojs/facial-landmarks-recognition/master/shape_predictor_68_face_landmarks.dat")

with open("shape_predictor_68_face_landmarks.dat", "wb") as f:
  f.write(out.content)
```

```
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

```
detector = dlib.get_frontal_face_detector()
```

```
dets = detector(img, 1)
```

```
dets
```

```
shape = predictor(img, dets[0])
```

```
points = shape.parts()
```

```
dir(points[0])
```

```
for point in points:
  img = cv2.circle(img, (point.x, point.y), 3, (255, 0, 0), 2)
```

```
plt.imshow(img)
```
